```python
def is_prime_simple(n):
    
    if n == 1:
        return True
    
    d = 2
    while n % d != 0:
        d += 1
    return d == n


def is_prime_squared(n):
    
    if n % 2 == 0:
         return n == 2
        
    limit = int(math.sqrt(n))
    d = 3
    while d <= limit and n % d != 0:
        d += 2
        
    return d * d > n


def is_prime_Fermat(n, n_tests=3):
    """
    a ** (n - 1) mod n = 1, а < n, если не выполняется, то число сложеное
    """
    
    if n == 1 or n == 2:
        return True
    
    res = []
    for i in range(n_tests):
        a = random.randint(1, n - 1)
        # a ** (n - 1) % n == 1
        if pow(a, n - 1, n) == 1:  # в степень по модулю
            # print("правдоподобно простое")
            res.append(True)
        else:
            res.append(False)
            # print("составное")
        if not all(res):
            return False
    
    return True
```

## Compare functions

```python
%%time

min_number = 1190_000
max_number = 1200_000
step = 1
numbers = list(range(min_number, max_number, step))

func1_time = []
func2_time = []
func3_time = []

for i in numbers:
    start = time.clock()
    _ = is_prime_simple(i)
    func1_time.append(time.clock() - start)
    
    start = time.clock()
    _ = is_prime_squared(i)
    func2_time.append(time.clock() - start)
    
    start = time.clock()
    _ = is_prime_Fermat(i)
    func3_time.append(time.clock() - start)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16, 6))
  
# time for simple func1 is much longer than others    
# ax.plot(numbers, func1_time, color="green", linewidth=4, ls="-", label="Simple func")
ax.plot(numbers, func2_time, color="red", linewidth=3, ls="-", label="Optimized func")
ax.plot(numbers, func3_time, color="blue", linewidth=2, ls="-", label="Ferma test")
ax.set(title="Functions execution time comparison", xlabel="numbers", ylabel="time (s)")
ax.legend(fontsize=16, loc=0)
ax.grid()

plt.subplots_adjust(wspace=0.15, hspace=0.27)
plt.show()
```
